Q1 - Import the life expectancy data into a dataframe (create the Jupyter notebook in Google Colab). Be sure to name the dataframe columns accordingly and sanitize the data as you did in SQL.

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# store the URL in url as parameter for requests
url = "https://raw.githubusercontent.com/bforoura/IDS/main/HW3/Life_Expectancy.csv"

# converting the CSV response to a dataframe
dataDF = pd.read_csv(url, sep=",")

#Fill in any missing values with the average value from each column
dataDF['Life_Expectancy '] = dataDF['Life_Expectancy '].fillna(dataDF['Life_Expectancy '].mean())
dataDF['Adult_Mortality'] = dataDF['Adult_Mortality'].fillna(dataDF['Adult_Mortality'].mean())
dataDF['Alcohol'] = dataDF['Alcohol'].fillna(dataDF['Alcohol'].mean())
dataDF['Percentage_Expenditure'] = dataDF['Percentage_Expenditure'].fillna(dataDF['Percentage_Expenditure'].mean())
dataDF[' BMI '] = dataDF[' BMI '].fillna(dataDF[' BMI '].mean())
dataDF['Total_Expenditure'] = dataDF['Total_Expenditure'].fillna(dataDF['Total_Expenditure'].mean())
dataDF['GDP'] = dataDF['GDP'].fillna(dataDF['GDP'].mean())
dataDF['Population'] = dataDF['Population'].fillna(dataDF['Population'].mean())
dataDF['Schooling'] = dataDF['Schooling'].fillna(dataDF['Schooling'].mean())


dataDF
````

Q2.1 - How does Adult mortality rates affect life expectancy. To solve this we can plug the data into a linear regression to see if there is any reasonable connection between the two values.

````python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Save the first column (all the rows) in X
X = dataDF.iloc[:, 3].values.reshape(-1,1)

# Save the second column (all the rows) in Y
Y = dataDF.iloc[:, 2].values.reshape(-1,1)

# Create the model
LRmodel = LinearRegression()

# Perform linear regression
LRmodel.fit(X, Y)

# Make predictions
Y_pred = LRmodel.predict(X)


# Visualize the dataset and the regression line:
plt.scatter(X, Y)
plt.plot(X, Y_pred, color='red')

plt.show()

# Error analysis
print("MSE: ", mean_squared_error(Y, Y_pred))
print("R2: ", r2_score(Y, Y_pred))
````

By the chart above it seems that that the higher level of adult mortality rate that exist the shorter life expectancy a nation can expect. The R2 value of 0.57 means that 57% of the variance of the dependent variable (Life Expectancy) being studied is explained by the variance of the independent variable (Adult Mortality Rate).
