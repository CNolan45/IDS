4.Create a new variable, saturated_ fat_ per_gram, by dividing the amount of saturated fat  by the weight in grams. Sort the data set by saturated_ fat_ per_gram and produce a listing of the five food  items highest in saturated fat per gram. Which food has the most saturated fat per gram?

```python
training_data['saturated_fat_per_gram'] = training_data['saturated_fat'] / training_data['weight_in_grams']

training_data.sort_values(by=['saturated_fat_per_gram'], ascending=False).iloc[0:5]
```

5.Standardize (zero mean) the field saturated_ fat_ per_gram, and produce a listing of all the food items that are outliers at the high end of the scale. How many food items are outliers at the low end of the scale? Create a density plot for saturated_ fat_ per_gram

```python
from scipy import stats

training_data['scaled_saturated_fat_per_gram'] = stats.zscore(training_data.saturated_fat_per_gram, nan_policy='omit')

training_data_outliers_high = training_data.query('(scaled_saturated_fat_per_gram > 3)')['scaled_saturated_fat_per_gram']
training_data_outliers_low = training_data.query('scaled_saturated_fat_per_gram < -3')['scaled_saturated_fat_per_gram']

print("The high outliers are\n",training_data.query('(scaled_saturated_fat_per_gram > 3)')['scaled_saturated_fat_per_gram'])
print("There are",training_data_outliers_high.count(), "high outliers")
print("There are",training_data_outliers_low.count(), "low outliers")
```
